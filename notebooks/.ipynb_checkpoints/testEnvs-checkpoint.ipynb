```python
import gym
from gym import envs

allNames = [str(k)[8:-1]  for k in envs.registry.all()]

print('| name | state space | action space | ')
print('|------|-------------|--------------|')
for name in allNames:
    try:
        env = gym.make(name)
        env.reset()
        observation, reward, done, info = env.step(env.action_space.sample())
        
        obsShape = str(observation.shape)
        
        env.close()
        print(f'| {name.rjust(40)} | {obsShape.rjust(15)}  | {env.action_space} |')
    except:
        pass
    
```
